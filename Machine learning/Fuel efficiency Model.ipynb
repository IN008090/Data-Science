## Draw a pie chart for distribution of car model 

```python
new_df['STANDARD TYPE'].unique()
```

```python
new_df['STANDARD TYPE'].value_counts()
```

```python
labels='Sedan','Van','SUV','Pick-up'
new_df['STANDARD TYPE'].astype(str).value_counts().plot(kind='pie',
                                                    figsize=(15,6),
                                                autopct='%1.1f%%',
                                                startangle=90,
                                                shadow=True,
                                                labels=['Sedan','Van','SUV','Pick-up']
                                                       )
plt.title('Sales as per car company ',y=1.12)
plt.axis('equal')
plt.legend(labels=labels,loc='upper left')
plt.show()
```

```python
new_df['MAKE'].value_counts(str)
```

```python
labels='Ford','Toyota','Chevrolet','Dodge','Nissan','GMC','Honda','Freightliner'
new_df.MAKE.astype(str).value_counts().plot(kind='pie',
                                         figsize=(20,10),
                                                autopct='%1.1f%%',
                                                startangle=90,
                                                shadow=True,
                                                labels=['Ford','Toyota','Chevrolet','Dodge','Nissan','GMC','Honda','Freightliner']
                                        )
plt.title('Car Makers distribution ',y=1.12)
plt.axis('equal')
plt.legend(labels=labels,loc='upper left')
plt.show()                                                      
```
