```python
# Importing the libaries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
digits= load_digits()
```

```python
dir(digits)  # This shows the properties of the in built data set
```

```python
df=pd.DataFrame(digits.data)
```

```python
# Now we need to visualized the data into matplotlib
%matplotlib inline 
plt.gray()
for i in range(9):
    plt.matshow(digits.images[i])  # Syntax to show the load_digits image via matplot lib we need to pass a for loop to run the range 
```

```python
digits.data[:7]
```

```python
df
```

```python
df['target']=digits.target
df
```

```python
X=df.drop('target', axis=1)
X
```

```python
y=df.target
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.2)
```

```python
len(X_train)
```

```python
len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier # Its is used to predict multiple outcomes following used in random forest algo
wed= RandomForestClassifier(30)
```

```python
wed.fit(X_train,y_train)
```

```python
wed.score(X_test,y_test)
```

```python
y_predictation=wed.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
dm=confusion_matrix(y_test,y_predictation)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(dm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')  # Success!
```
