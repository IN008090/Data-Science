### Axis, label, grid in graph representation 

```python
import matplotlib.pyplot as plt
%matplotlib inline  
from IPython.core.display import HTML
```

```python
days =[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")
```

```python
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.ylabel('Temp range')
plt.xlabel('sample data ')
plt.title("Min and Max weather temp chart")
plt.legend()  # To present the 'max','min','avg' specification 
plt.legend(loc="upper right")  # This help to locate the legend in the graph
plt.legend(loc="best")  # Automatically decide where to locate
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()  # To place the grid in the graph
# These all label, legend, grid all will come in one cell it will not run indivisually 
plt.savefig('Weather_graph.png')
```

```python
plt.bar(days,max_t,label="Max")
plt.bar(days,min_t,label="Min")
plt.ylabel('Temp range')
plt.xlabel('sample data ')
plt.title("Min and Max weather temp chart")
plt.savefig('Min & Max.png',pad_inches=30,edgecolor='white',transparent='facecolor')
```
