### Matplotlib is used present the data in a graphically method 

```python
import matplotlib.pyplot as plt 
%matplotlib inline # This is important to write along
```

```python
x=[1,2,3,4,5,6,7] # x have no of week days 
y=[50,51,52,48,47,49,46]# y have highest no of temp.
```

```python
plt.xlabel('Day')  # To label x, y along with title can use plt. 
plt.ylabel('Temperature')
plt.title('weather')
```

```python
plt.plot(x,y) # This method is used to plot the graph between x and y gragh
```

```python
plt.plot(x,y,color='green') # Method to change the color of the plot
```

```python
plt.plot(x,y,color='green',linewidth=5) # To increase the width of the plot 
```

```python
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted') # To represent the plot in dot sign
```

```python
plt.plot(x,y,color='red',linewidth=3, linestyle='--')
```
