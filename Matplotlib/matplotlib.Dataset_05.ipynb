## Hist() 

```python
# Importing libaries
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline 
market_price=[354,783,874,387,783,543,354]
market_ramp=[360,790,880,390,790,550,360]
plt.title('Market research table')
plt.xlabel('market_price')
plt.ylabel('frequency')

plt.hist([market_price,market_ramp],bins=[400,550,800,900],color=['red','green'],label=['market_price','frequency'],rwidth=.65)
```

```python
plt.pie(market_price,labels=['Parle','Lotto choco pie','Mangovite','Bread','Butter','Coconut water','Chocolate'],shadow=True,
        frame=False,pctdistance=1,radius=2,explode=[0,0.4,0,0,0,0,0],autopct='%0.2f%%')
plt.show()

```
