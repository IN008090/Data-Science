## How to save matplot chart in image or PDF files

```python
# Importing the libaries
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Importing the csv file 
df=pd.read_csv("weather_grao.csv")
```

```python
df
```

```python
new_df=df.loc[:5]
new_df
```

```python
new_f=new_df['Stn Press (kPa)']
new_f
```

```python
%matplotlib inline 
plt.pie(new_f,radius=2,shadow=2,explode=[0,0,0,0,0,0],labels=['Atmospheric pressure 1','Atmospheric pressure 2','Atmospheric pressure 3','Atmospheric pressure 4','Atmospheric pressure 5','Atmospheric pressure 6'],labeldistance=1.02)
plt.show()  
plt.savefig("Atmospheric.png",pad_inches=2,bbox_inches='tight')
```
